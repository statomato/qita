```python
import pandas as pd

academic_num = pd.read_csv('academic_num.csv')
```

```python
life_law = pd.read_csv("life_law.csv")
life_law.head(10)
```

```python
col = ['d1', 'd2']
life_law.columns = col
life_law['d1']=life_law['d1'].fillna("")
life_law.head(1)
```

```python
def ee(col) :
    col = col.replace('▪ ','')
    col = col.replace(' ','')
    return col.split("(")[0]
```

```python
life_law['d1_re'] = life_law['d1'].apply(ee)
life_law['d2_re'] = life_law['d2'].apply(ee)
life_law.head(3)
```

```python
life_law1 = life_law[['d1','d1_re']]
life_law2 = life_law[['d2','d2_re']]
```

```python
merge_col = academic_num[['학수번호','교과목명']]
life_law1 = pd.merge(life_law1, merge_col, left_on='d1_re', right_on='교과목명', how='left')
life_law2 = pd.merge(life_law2, merge_col, left_on='d2_re', right_on='교과목명', how='left')
```

```python
drop_col1 = ['d1_re','교과목명']
drop_col2 = ['d2_re','교과목명']

life_law1.drop(drop_col1, axis=1, inplace=True)
life_law2.drop(drop_col2, axis=1, inplace=True)
```

```python
def ee2(col) :
    col = col.replace('▪ ','')
    return col
```

```python
life_law1['d1'] = life_law1['d1'].apply(ee2)
life_law2['d2'] = life_law2['d2'].apply(ee2)
```

```python
life_law1 = life_law1.dropna(axis=0)
life_law1
```

```python
life_law2
```

```python
life_law1['result']=life_law1['d1']+'('+life_law1['학수번호']+')'
life_law2['result']=life_law2['d2']+'('+life_law2['학수번호']+')'
```

```python
life_law1.head(5)
```

```python
life_law2.head()
```

```python
life_law1['result'].to_csv("life_law1.csv",index=False)
life_law2['result'].to_csv("life_law2.csv",index=False)
```

```python
def make_df(df):
    col = ['d1', 'd2']
    df.columns = col
    df['d1']=df['d1'].fillna("")
    
    df['d1_re'] = df['d1'].apply(ee)
    df['d2_re'] = df['d2'].apply(ee)
    
    df1 = df[['d1','d1_re']]
    df2 = df[['d2','d2_re']]
    
    merge_col = academic_num[['학수번호','교과목명']]
    df1 = pd.merge(df1, merge_col, left_on='d1_re', right_on='교과목명', how='left')
    df2 = pd.merge(df2, merge_col, left_on='d2_re', right_on='교과목명', how='left')
    
    drop_col1 = ['d1_re','교과목명']
    drop_col2 = ['d2_re','교과목명']
    
    df1.drop(drop_col1, axis=1, inplace=True)
    df2.drop(drop_col2, axis=1, inplace=True)
    
    df1['d1'] = df1['d1'].apply(ee2)
    df2['d2'] = df2['d2'].apply(ee2)
    
    df1['result']=df1['d1']+'('+df1['학수번호']+')'
    df2['result']=df2['d2']+'('+df2['학수번호']+')'
    
    return df1, df2
```
